### Reflection on Project 

<b>Lane Finding Pipeline</b>

The current pipeline converts input images to grayscale, applies a gaussian smoothing operation with a kernel size of 5, then applies canny edge detection to find edges. The image is then masked to identify only those edges that are with a triangular shaped region directly in front of the car. The canny edges are passed to a hough transform function which identifies lines which are more than 100 pixels long, but allows gaps of 100 pixels between points in the line. The resulting lines from the hough transform are averaged together, and then extrapolated to produce the resulting lane lines. <b>Shortcomings and Improvements</b>

When the pipeline was applied to the video, the lane lines are very shaky, and occasionally in the wrong place. However, the majority of the time, the lane lines are correct.

The pipeline could be made more robust by spending more time tweaking the parameters to select lane lines in a more robust manner. 

The pipeline expects to see straight lines, two on either side, meeting at a vanishing point near the centre of the image. If these conditions are not satisfied, the pipeline can fail.

The pipline can fail if it finds more than two lines, say if there was a pavement, or cracks in the road, that it identifies as lines. This could be improved by identifying lots of lines, representing things such as lane lines, signs, cracks etc, then selecting the ones that are most likely lane lines.

The pipeline can fail if the road is too curved, since it is trying to find straight lines. This could be improved by trying to fit curved lines to the road, or multiple short line segments, rather than one single straight line.
